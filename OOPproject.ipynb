```python
import nlpcloud
class MyApp :
  def __init__(self):
    self.__database = {}
    self.__first_menu()
  def __first_menu(self):
    first_input= input("""
            Hi There!
          1.New Here ? Register
          2.Already a member ? Login
          3.wanna leave? exit
    """)
    if first_input == '1' :
      self.__register()
    elif first_input == '2' :
      self.__login()
    else :
      exit()
  def __register(self):
    name = input('enter your name ')
    email = input('enter your email ')
    password = input('create your password ')
    if email in self.__database:
      print('already registered. go to login!!')
      self.__login()
    else :
      self.__database[email]=[name,password]
      print('now login to your account')
      self.__login()
  def __login(self):
    email = input('enter your registered mail ')
    password = input('enter your password ')
    if email not in self.__database :
      print('email not registered!!')
      self.__first_menu()
    else :
      if password == self.__database[email][1] :
        print('login successful!!')
        self.__second_menu()
      else:
        print('wrong password re enter credentials')
        self.__login()
  def __second_menu(self):
    second_input = input("""
            Hi What would you like to do!
          1.show intent
          2.generate header
          3.code generator
          4.exit
    """)
    match second_input :
      case '1':
        self.__intent()
      case '2':
        self.__header()
      case '3':
        self.__coder()
      case '4':
        exit()
  def __intent(self):
    text = input('enter text whose intent you need to find')
    client = nlpcloud.Client("finetuned-llama-3-70b", "9eca2121102077306269e9ef6525da6a2f787db9", gpu=True)
    print('\n',client.intent_classification(text)['intent'])
    self.__second_menu()
  def __header(self):
    text = input('enter text whose header is to be generated ')
    client = nlpcloud.Client("t5-base-en-generate-headline", "9eca2121102077306269e9ef6525da6a2f787db9", gpu=False)
    print('\n',client.summarization(text)['summary_text1'])
    self.__second_menu()
  def __coder(self):
    prompt = input('enter prompt to generate code')
    client = nlpcloud.Client("finetuned-llama-3-70b", "9eca2121102077306269e9ef6525da6a2f787db9", gpu=True)
    print(client.code_generation(prompt)['generated_code'])
    self.__second_menu()
ob = MyApp()
```
